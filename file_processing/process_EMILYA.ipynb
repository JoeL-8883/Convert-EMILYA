### Copies EMILYA files into more organised folders suitable for retraining

```python
# Careful not to run this twice otherwise getcwd() will be the location of All_BvH_Files
current = os.getcwd()
parent = os.path.dirname(os.path.dirname(os.getcwd()))
```

```python
def adverb(emotion):
    if emotion == 'Anger':
        return 'angrily'
    elif emotion == 'Anxiety':
        return 'anxiously' # nervously
    elif emotion == 'Joy': 
        return 'joyfully' # cheerfully, cheerily, happily
    elif emotion == 'Neutral':
        return ''
    elif emotion == 'Panic Fear':
        return 'panicked' # fearfully, scaredly, panically
    elif emotion == 'Pride':
        return 'proudly' # boastfully
    elif emotion == 'Sadness':
        return 'sadly' # sorrowfully, unhappily, miserably
    elif emotion == 'Shame':
        return 'shamefully ' # guiltily, embarrassedly

def describe_motion(motion, emotion, texts_dir, number):
    if motion == 'BS':
        pass
    elif motion == 'CS':
        pass
    elif motion == 'KD':
        pass
    elif motion == 'Lf':
        pass
    elif motion == 'MB':
        pass
    elif motion == 'SD':
        pass
    elif motion == 'SDBS':
        pass
    elif motion == 'SW':
        pass
    elif motion == 'Th':
        pass
    elif motion == 'TrSW':
        pass
    elif motion == 'TrWH':
        pass
    elif motion == 'WH':
        pass

    # Write generic description based on emotion
    if emotion == 'Anger':
        with open(os.path.join(texts_dir, number + '.txt'), 'w') as f:
            f.write('This is an angry motion')
    elif emotion == 'Anxiety':
        with open(os.path.join(texts_dir, number + '.txt'), 'w') as f:
            f.write('This is an anxiety motion')
    elif emotion == 'Joy':
        with open(os.path.join(texts_dir, number + '.txt'), 'w') as f:
            f.write('This is a pride motion')
    elif emotion == 'Neutral':
        with open(os.path.join(texts_dir, number + '.txt'), 'w') as f:
            f.write('This is a joy motion')
    elif emotion == 'Panic Fear':
        with open(os.path.join(texts_dir, number + '.txt'), 'w') as f:
            f.write('This is a sadness motion')
    elif emotion == 'Pride':
        with open(os.path.join(texts_dir, number + '.txt'), 'w') as f:
            f.write('This is a panic fear motion')
    elif emotion == 'Sadness':
        with open(os.path.join(texts_dir, number + '.txt'), 'w') as f:
            f.write('This is a shame motion')
    elif emotion == 'Shame':
        with open(os.path.join(texts_dir, number + '.txt'), 'w') as f:
            f.write('This is a neutral motion')
```

```python
def create_output_dir(current_dir, input_dir, copy_bvh=False, name_code=False, texts=False, raw=False): 
    counter = 0
    # input_dir = the directory of All_BvH_Files i.e. ../experiments/All_BvH_Files
    outputdir = current_dir # the dir of this notebook
    outputdir_categorised = os.path.join(outputdir, 'output_categorised')
    output_dir_raw = os.path.join(outputdir, 'output_raw')
    if not os.path.exists(output_dir_raw):
        os.makedirs(output_dir_raw)

    # Get the motions in the input directory
    emilya_motions = os.listdir(input_dir) # the different types of motions, i.e. SW
    
    # Iterate through each of the motions, SW, BS, CS, etc.
    for motion in emilya_motions:       
        # Create folders for each type of motion
        output_motion_folders = os.path.join(outputdir_categorised, motion) # i.e. ../output/SW
        if not os.path.exists(output_motion_folders):
            os.makedirs(output_motion_folders) 
        
        # Get the emotions for each motion
        if motion.endswith('.DS_Store'):
            continue
        else:
            emotions = os.listdir(motion)

            # Write each emotion folder to output directory
            for emotion in emotions[1:]:
                output_emotion_folder = os.path.join(output_motion_folders, emotion) # i.e. ../output/SW/angry
                if not os.path.exists(output_emotion_folder):
                    os.makedirs(output_emotion_folder)

                # Copy the BVH files to the output directory if specified
                if copy_bvh:
                    # Get each of the actors mocap data for a motion and emotion
                    actors_dir = os.path.join(input_dir, motion, emotion) # i.e. ../experiments/All_BvH_Files/SW/angry
                    actors = os.listdir(actors_dir)

                    # Get the directory of the actors mocap data containg BVH files
                    for actor in actors:
                        bvh_folder = os.path.join(actors_dir, actor)
                        if bvh_folder.endswith('.DS_Store'): 
                            continue
                        else:
                            bvh_files = os.listdir(bvh_folder)

                            # Copy the BVH files to the output directory
                            for bvh in bvh_files:
                                bvh_file = os.path.join(actors_dir, actor, bvh)
                                number = f"{counter:06d}"
                                final_destination_raw = os.path.join(output_dir_raw, number + '.bvh')
                                final_destination = os.path.join(output_emotion_folder, number + '.bvh')

                                if name_code: 
                                    if raw: # Just throw all the .bvh files into one folder
                                        shutil.copy(bvh_file, final_destination_raw)

                                        # If the texts are true then write a generic description of the motion with emotion
                                        if texts:
                                            # Create a directory for text files
                                            texts_dir = os.path.join(outputdir, 'texts')
                                            if not os.path.exists(texts_dir):
                                                os.makedirs(texts_dir)
                                                describe_motion(motion, emotion, texts_dir, number) # Don't touch this for now
                                    else:
                                        shutil.copy(bvh_file, final_destination)
                                else:
                                    shutil.copy(bvh_file, output_emotion_folder)
                                counter += 1

                    

# Get the directory of the BvH files
bvh_files = os.path.join(parent, 'All_BvH_Files')
try:
    os.chdir(bvh_files)
except:
    raise FileNotFoundError(f"The directory {bvh_files} does not exist.")

create_output_dir(current, bvh_files, copy_bvh=True, name_code=True, texts=False, raw=True)
```
